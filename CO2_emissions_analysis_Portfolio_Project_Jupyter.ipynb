## Repartition of the country based on their CO2 emissions - comparison between 1750 and 2022
(Lots of missing value for the first years)

```python
#Select the rows for each country or region corresponding to 1750
repartition_1750 = df.loc[df['year'] == 1750]
```

```python
repartition_1750.shape
```

```python
plt.style.use('seaborn-v0_8')
chart = sns.histplot(data=repartition_1750['co2'])
chart.set_title('Repartition of country based on their CO2 emissions in 1750')
chart.set_xlabel('CO2 emissions in millions tons')
chart.set_ylabel('Count - Number of countries')
plt.show()
```

```python
#Select the rows for each country or region corresponding to 2022
repartition_2022 = df.loc[df['year'] == 2022]
```

```python
repartition_2022.shape
```

```python
plt.style.use('seaborn-v0_8')
chart = sns.histplot(data=repartition_2022['co2'])
chart.set_title('Repartition of country or region based on their CO2 emissions in 2022')
chart.set_xlabel('CO2 emissions in millions tons')
chart.set_ylabel('Count - Number of countries')
plt.show()
```

```python
#Some very high values make the histogram difficult to read
#These values correspond to World, Group of country (OECD...), Continent (Asia...) or the countries emmiting the most CO2
repartition_2022.sort_values('co2',ascending=False)
```

```python
repartition_2022['co2'].max()
```

```python
#Zoom excluding the largest value (World, Continent and Region) 
plt.style.use('seaborn-v0_8')
chart = sns.histplot(data=repartition_2022['co2'])
chart.set_xlim(-5,3100)
chart.set_title('Repartition of country or region based on their CO2 emissions in 2022')
chart.set_xlabel('CO2 emissions in millions tons')
chart.set_ylabel('Count - Number of countries')
plt.show()
```

Lots of values are missing for the year 1750 mostly because of lack of records (some country also didn't exist at the time). We can look at an earliest year like 1900 to compare it to 2022. 
In the most recent years, some countries are responsible for a large part, of the CO2 emissions like China. In the past the repartition was more uniform meaning that a larger group of country was contributing to the emissions opposed to some individual country nowadays.

## Top 10 Countries in CO2 emission from 1750 to 2022

```python
plt.style.use('seaborn-v0_8')
chart = df.groupby('country')['co2'].sum().sort_values(ascending=False)[1:11].plot(kind='bar')
chart.set_title('Top 10 Countries in CO2 emission from 1750 to 2022')
chart.set_ylabel('CO2 Emissions in million ton')
```
